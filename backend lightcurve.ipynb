```python
filename = "/home/swathi/data/20151016_P01_126T01_9000000030_level2/SSM/SSM_00273/evtlstgtifil/AS1P01_126T01_9000000030_00273ssm_level2_clean_ssm1.fits"
fits.info(filename)
```

```python
with fits.open(filename) as hdulist: 
    header1 = hdulist[1].header
  
print(repr(header1[0:24])) #repr() prints the info into neat columns
```

```python
with fits.open(filename) as hdulist:
    binaryext = hdulist[1].data

binarytable = Table(binaryext)
binarytable[1:5]
```

```python
with fits.open(filename, mode="readonly") as hdulist:
    # Read in the "BJDREF" which is the time offset of the time array.
    bjdrefi = hdulist[1].header['TTYPE1'] 
    bjdreff = hdulist[1].header['TTYPE6']

    # Read in the columns of data.
    times = hdulist[1].data['SSM_PL_EVTTIME_MJD'] 
    sap_fluxes = hdulist[1].data['SSM_PL_EVTLV']
    #pdcsap_fluxes = hdulist[1].data['SSM_PL_EVTRV']
```

```python
# Convert the time array to full BJD by adding the offset back in.
bjds = bjdrefi + bjdreff 

plt.figure(figsize=(9,4))

# Plot the time, uncorrected and corrected fluxes.
plt.plot(times, sap_fluxes, '-k', label='SSM_PL_EVTLV') 
#plt.plot(bjds, pdcsap_fluxes, '-b', label='SSM_PL_EVTRV') 

plt.title('SSM Light Curve')
plt.legend()
plt.xlabel('Time (days)')
plt.ylabel('Flux (electrons/second)')
plt.show()
```
